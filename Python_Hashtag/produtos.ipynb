```python
import pyautogui as pa
import time
import pandas as pd
```

```python
link = 'https://www.youtube.com/'
pa.PAUSE = 2.5
```

```python
# Acessando o Firefox e o youtube
pa.press('win')
pa.write('firefox')
pa.press('enter')
pa.write(link)
pa.press('enter')
```
