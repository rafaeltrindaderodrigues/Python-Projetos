```python
import pyautogui as pa
import pandas as pd
import time
```

```python
df = pd.read_csv('dados_fraldas.csv')
```

```python
display(df)
```

```python
# Fazendo automação de abrir o site 

# Definir um tempo mínimo para não ser tão rápido
pa.PAUSE = 3

pa.press('win')    

# Buscando pelo firefox e pesquisando pelo link do lovable
link = 'https://receita-facil-bot.lovable.app'
pa.write('chrome')
pa.press('enter')
pa.write(link, interval=0.05)
pa.press('enter')

```

```python
# A partir de agora eu faço a interação com cada item na tabela

for i, row in df.iterrows():
    pa.click(x=828, y=425)
    pa.write(row.loc['nome_usuario'], interval=0.05)
    pa.press('tab')
    pa.write(str(row.loc['cpf_usuario']), interval=0.05)
    pa.press('tab')
    pa.write(str(row.loc['crm_medico']), interval=0.05)
    pa.press('tab')
    pa.write(str(row.loc['data_receita']), interval=0.05)
    pa.press('tab')
    pa.write(row.loc['tamanho_fraldas'], interval=0.05)
    pa.click(x=962, y=851)
```
